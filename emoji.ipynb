```python
#data was separated by tab.
japan_emoji= pd.read_csv ('japanD.txt', sep = '\t')

japan_emoji.head()
```

```python
#Combining Korean emoji and Japanese emoij as East Asian emoji
East_Asia = [korea_emoji, japan_emoji]
East_Asia_emoji = pd.concat(East_Asia)

#including one variable as culture and assign value 1 for t-test
East_Asia_emoji.insert (1, "culture", 1, True)

#creating one variable as emojiSum that combines all emoticons.
East_Asia_emoji_sum = East_Asia_emoji.loc[:,'emojiPersonalPositive':'emoticonNegative'].sum(axis=1)
East_Asia_emoji.insert(4, "emojiSum", East_Asia_emoji_sum)
East_Asia_emoji.tail()
```

```python
#combining US emoji and Canada emoiji
North_America = [us_emoji, canada_emoji]
North_America_emoji = pd.concat(North_America)

#including one variable as culture and assign value 2 for t-test
North_America_emoji.insert (1, "culture", 2, True)

#creating one variable as emojiSum that combines all emoticons.
North_America_emoji_sum = North_America_emoji.loc[:,'emojiPersonalPositive':'emoticonNegative'].sum(axis=1)
North_America_emoji.insert (4, "emojiSum", North_America_emoji_sum)

North_America_emoji.head()
```

```python
both_culture = [East_Asia_emoji,North_America_emoji]
both_culture_emoji = pd.concat(both_culture)
both_culture_emoji.tail()
```

```python
#running welch-t test rather than student t due to unequal variance
from scipy import stats

#stats.ttest_ind(East_Asia_emoji['emojiPersonalPositive'], North_America_emoji['emojiPersonalPositive'], equal_var = False)

stats.ttest_ind(East_Asia_emoji['emojiSum'], North_America_emoji['emojiSum'], equal_var = False)
```

```python
#from https://pythonfordatascienceorg.wordpress.com/welch-t-test-python-pandas/.

def welch_dof(x,y):
        dof = (x.var()/x.size + y.var()/y.size)**2 / ((x.var()/x.size)**2 / (x.size-1) + (y.var()/y.size)**2 / (y.size-1))
        print(f"Welch-Satterthwaite Degrees of Freedom= {dof:.4f}")
        
def welch_ttest(x, y): 
    ## Welch-Satterthwaite Degrees of Freedom ##
    dof = (x.var()/x.size + y.var()/y.size)**2 / ((x.var()/x.size)**2 / (x.size-1) + (y.var()/y.size)**2 / (y.size-1))
   
    t, p = stats.ttest_ind(x, y, equal_var = False)
    
    print("\n",
          f"Welch's t-test= {t:.4f}", "\n",
          f"p-value = {p:.20f}", "\n",
          f"Welch-Satterthwaite Degrees of Freedom= {dof:.4f}")
    
 
```

```python
#welch_ttest(East_Asia_emoji['emojiPersonalPositive'], North_America_emoji['emojiPersonalPositive'])
welch_ttest(East_Asia_emoji['emojiSum'], North_America_emoji['emojiSum'])
```
